```python
#libraries
import numpy as np
from netCDF4 import Dataset
import subprocess
from pathlib import Path
from scipy.stats import mstats
#model output file
model_ncf='../examples/run_dir/RiceUSTWT/riceUStwt.ecosim.h0.2011-01-01-00000.nc'
model_ncf='../examples/run_dir/blodgett/Blodget.ctrl.ecosim.h0.2001-01-01-00000.nc'

NZ=0
print('%-20s[%-26s],%-15s,%-15s,%-15s,%60s'%('variable','unit','min','median','max','long_name'))
print('-'*100)
with Dataset(model_ncf, 'r+') as nc_file:
    
    Variable_List = ['JMax25C_photo_pft','SLA_pft','VcMax25C_RUBISCO_pft','VoMax25C_RUBISCO_pft',
                     'VmaxNH4Root_pvr','VmaxNO3Root_pvr','LeafProteinC_pft','LeafRubisco_pft',
                    'LeafChlC_pft','CAN_HT_pft','LAIstk_pft']
    for var in Variable_List:
        myvar=nc_file.variables[var]
        var_dat=myvar[:]
        units=myvar.units
        long_name=myvar.long_name
        ndim=len(var_dat.shape)-1        
        if ndim==1:
            dat=var_dat[:,NZ]
        elif ndim==2:
            dat=var_dat[:,:10,NZ].reshape(-1)
        #remove small values  
        masked_data = np.ma.masked_less(dat, 1.e-5).compressed()        
        typival=np.percentile(masked_data, [2.5, 50, 97.5])        
        print('%-20s[%-26s],%-15.6f,%-15.6f,%-15.6f,%-60s'%(var,units,typival[0],typival[1],typival[2],long_name))
            
#    return percentiles = )
```
