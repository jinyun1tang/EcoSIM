```python
import numpy as np
from netCDF4 import Dataset
from datetime import datetime

current_dateTime = datetime.now()
nc_f='../input_data/MicrobePars.%4d%02d%02d.nc'%(current_dateTime.year,current_dateTime.month,current_dateTime.day)

nc_fid = Dataset(nc_f, 'w')

nc_fid.description='Microbial parameters created on %4d/%02d/%02d/%02d:%02d:%02d'% \
(current_dateTime.year,current_dateTime.month,current_dateTime.day, \
 current_dateTime.hour,current_dateTime.minute,current_dateTime.second)

for key,val_str in  microbe_pars.items():    
    parts = val_str.rsplit(':', 1)
    description_with_units = parts[0]
    vals=parts[1].strip()
    if ',' in vals:
        val=vals.rsplit(',')
        ndim=len(val)
    else:
        val=vals
        ndim=1         
    parts=description_with_units.rsplit('[',1)
    description = parts[0].strip()
    unit = parts[1].rstrip(']').strip()
    if unit=='dimension':
        """
        create dimension
        """
        nc_fid.createDimension(key, int(val))        
    else:
        """
        create variable
        """
        if '(' in key:            
            """array"""
            parts=key.split('(')            
            dimstr = parts[1].rstrip(')')
            if ',' in dimstr:
                dims=dimstr.split(',')
            else:
                dims=dimstr
            print(f'{parts[0]} {dims}')
            w_nc_var = nc_fid.createVariable(parts[0], 'f4',(dims))
            w_nc_var.long_name=description
            w_nc_var.units=unit
            arr=[float(item) for item in val]
            if len(dims) > 1:
                target_shape = w_nc_var.shape
                nc_fid.variables[parts[0]][:]=np.reshape(arr,target_shape)
            else:
                nc_fid.variables[parts[0]][:]=arr
        else:
            """scalar"""
            w_nc_var = nc_fid.createVariable(key, 'f4')
            w_nc_var.long_name=description
            w_nc_var.units=unit
            nc_fid.variables[key][:]=float(val)

nc_fid.close()  # close the new file
```
