# Labeld fertilizer
Typical fertilizer grades are expressed as three numbers separated by dashes, representing the percentages by weight of the three primary macronutrients: nitrogen (N), phosphorus (P2O5), and potassium (K2O). Some common fertilizer grades include:

10-10-10: Balanced fertilizer with equal percentages of N, P, and K for general purpose use.

15-15-15: Another balanced fertilizer commonly used to supply equal amounts of the major nutrients.

11-52-0 (Monoammonium Phosphate): High in phosphorus, with some nitrogen, no potassium.

18-46-0 (Diammonium Phosphate): High phosphorus and nitrogen.

30-0-0 to 46-0-0 (Urea): High nitrogen fertilizers with no phosphorus or potassium.

0-0-60 (Potassium Chloride): High potassium fertilizer.

13-0-44 (Potassium Nitrate): Supplies potassium and nitrogen.

0-44-0 to 0-52-0 (Superphosphate and Triple Superphosphate): High phosphorus fertilizers.

7-7-7 (Growmore): Balanced fertilizer used in various applications.

These grades vary depending on crop needs and soil fertility. Fertilizers may also include secondary nutrients like sulfur, calcium, and magnesium, or micronutrients such as iron, zinc, or boron, sometimes indicated by additional numbers.


# Detail grades
 **For the fertilizer grade 30-0-20**, which indicates 30% nitrogen (N), 0% phosphate (P2O5), and 20% potassium (K2O), the nitrogen fraction can come from different nitrogen forms such as ammonium (NH4+), nitrate (NO3-), and urea (CO(NH2)2).

**For the fertilizer 11-52-0**, which is Monoammonium Phosphate (MAP), the nitrogen content is almost entirely in the form of ammonium (NH4+). It contains about 11% nitrogen, all as ammoniacal nitrogen (NH4+), with no nitrogen as nitrate (NO3−) or urea. The chemical formula is NH4H2PO4, indicating the nitrogen is from the ammonium ion. There is no urea nitrogen in this fertilizer.

**The fertilizer 15-15-15** is a balanced NPK fertilizer containing equal proportions (15% each) of nitrogen (N), phosphorus (P2O5), and potassium (K2O). Regarding the forms of nitrogen in 15-15-15 fertilizers, these typically contain a balanced mixture of ammonium nitrogen (NH4+) and nitrate nitrogen (NO3−), with urea nitrogen generally not a major component.
