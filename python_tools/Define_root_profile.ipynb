```python
import matplotlib.pyplot as plt
import numpy as np
d=np.linspace(0, 200, 20)
beta=0.97
y=1.-beta**d
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))
ax1.plot(y,-d)
ax1.set_ylabel('Depth (cm)')
ax1.set_xlabel('Dumulative fine root fraction')
ax1.set_title('Jackson power law model of fine root profile')

ax2.plot(np.diff(y))
ax2.set_title('root distribution')
plt.show()
print(y)
```
